```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

```python
mu=2.0
sigma=0.5

norm_rv=sts.norm(loc=mu,scale=sigma)
norm_rv.rvs(size=10)
```

```python
norm_rv.cdf(3)
```

```python
x=np.linspace(0,4,100)
print x[-5:]
```

```python
cdf=norm_rv.cdf(x)
plt.plot(x,cdf)
plt.ylabel('$F(x)$')
plt.xlabel('$x$')
```

```python
norm_rv.pdf(3)
```

```python
x=np.linspace(0,4,100)
pdf=norm_rv.pdf(x)
plt.plot(x,pdf)
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
```

```python
a=1
b=4
uniform_rv=sts.uniform(a,b-a)
uniform_rv.rvs(10)
```

```python
x=np.linspace(0,5,100)
cdf=uniform_rv.cdf(x)
plt.plot(x,cdf)
plt.xlabel('$x$')
plt.ylabel('$F(x)$')
```

```python
x=np.linspace(0,5,100)
pdf=uniform_rv.pdf(x)
plt.plot(x,pdf)
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
```

```python
bernoulli_rv=sts.bernoulli(0.1)
bernoulli_rv.rvs(10)
```

```python
binomial_rv=sts.binom(20,0.7)
binomial_rv.rvs(10)
```

```python
x=np.linspace(0,20,21)
cdf=binomial_rv.cdf(x)
plt.step(x,cdf)
plt.xlabel('$x$')
plt.ylabel('$F(x$)')
```

```python
x=np.linspace(0,20,21)
pmf=binomial_rv.pmf(x)
plt.plot(x,pmf,'*')
plt.xlabel('$x$')
plt.ylabel('$P(X=x)$')
```

```python
x=np.linspace(0,45,46)
for N in [20,30]:
    for p in [0.2,0.7]:
        rv=sts.binom(N,p)
        cdf=rv.cdf(x)
        plt.step(x,cdf,label='$N=%s, p=%s' % (N,p))
plt.legend()
plt.title('CDF (binomial)')
plt.ylabel('$F(X)$')
plt.xlabel('$x$')
```

```python
symbols=iter(['o','s','^','+'])
for N in [20,30]:
    for p in [0.2,0.8]:
        rv=sts.binom(N,p)
        pmf=rv.pmf(x)
        plt.plot(x,pmf,next(symbols),label='$N=%s, p=%s' % (N,p))
plt.legend()
plt.title('PMF (binomial)')
plt.xlabel('$x$')
plt.ylabel('$P(X=x)$')
```

```python
poisson_rv=sts.poisson(5)
poisson_rv.rvs(10)
```

```python
x=np.linspace(0,30,31)
for l in [1,5,10,15]:
    rv=sts.poisson(l)
    cdf=rv.cdf(x)
    plt.step(x,cdf,label='$\lambda=%s$'%l)
plt.legend()
plt.title('CDF (poisson)')
plt.ylabel('$F(x)$')
plt.xlabel('$x$')
```

```python
x=np.linspace(0,30,31)

symbols=iter(['o','s','^','+'])
for l in [1,5,10,15]:
    rv=sts.poisson(l)
    pmf=rv.pmf(x)
    plt.plot(x,pmf,next(symbols),label='$\lambda=%s$'%l)
plt.legend()
plt.title('PMF (poisson)')
plt.xlabel('$x$')
plt.ylabel('$P(X=x)$')
```

```python
elements=np.array([1,5,12])
probabilities=[0.05,0.7,0.25]
np.random.choice(elements,10,p=probabilities)
```

```python
x=np.linspace(0,30,100)
for k in [1,2,3,4,6,9]:
    rv=sts.chi2(k)
    cdf=rv.cdf(x)
    plt.plot(x,cdf,label='$k=%s$'%k)
plt.legend()
plt.title('CDF ($\chi^2_k$)')
```

```python
x=np.linspace(0,30,100)
for k in [1,2,3,4,6,9]:
    rv=sts.chi2(k)
    pdf=rv.pdf(x)
    plt.plot(x,pdf,label='$k=%s$'%k)
plt.legend()
plt.title('PDF ($\chi^2_k$)')
```
